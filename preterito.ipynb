```python
import time

# el pretérito o el pretérito perfecto simple


def preterito():
    """
    El pretérito o el pretérito perfecto simple:
    - The pretérito is NOT used when we want to talk about
    continuous or habitual actions in the past with no specific
    beggining or end. In this case the pretérito imperfecto is used.

    - completed events
    - beginnings and ends
    - specific times and dates
    - events in sequence

    The regular conjugation of the 'pretérito' follows these rules:
        - remove the infinitive ending (i.e., -ar, -er, -ir)
        - if -ar verb:
            >  verb + ['é', 'aste', 'ó', 'amos', 'asteis', 'aron']
        - if -er/-ir verb:
            >  verb + ['í', 'iste', 'ió', 'imos', 'isteis', 'ieron']

    There are 4 irregular verbs in 'pretérito': ser, ir, dar, ver
    ser: fui, fuiste, fue, fuimos, fuisteis, fueron
    ir: fui, fuiste, fue, fuimos, fuisteis, fueron
    dar: di, diste, dio, dimos, disteis, dieron
    ver: vi, viste, vio, vimos, visteis, vieron
    """
    verb = str(input("Darme un infinitivo en Castellano:")).strip().lower()
    ser_ir_preterito = ["fui", "fuiste", "fue", "fuimos", "fuisteis", "fueron"]
    dar_preterito = ["di", "diste", "dio", "dimos", "disteis", "dieron"]
    ver_preterito = ["vi", "viste", "vio", "vimos", "visteis", "vieron"]
    ar_suffixes = ["é", "aste", "ó", "amos", "asteis", "aron"]
    er_ir_suffixes = ["í", "iste", "ió", "imos", "isteis", "ieron"]
    estar_tener_andar_verb = ["uve", "uviste", "uvo", "uvimos", "uvisteis", "uvieron"]
    caber_saber_poner_hacer_querer_decir_suf = [
        "e",
        "iste",
        "o",
        "imos",
        "isteis",
        "ieron",
    ]

    # IRREGULAR VERBS
    # ser and ir
    if verb == "ser" or verb == "ir":
        for case in ser_ir_preterito:
            print(case)
            time.sleep(1)
    elif verb == "dar":
        for case in dar_preterito:
            print(case)
            time.sleep(1)
    elif verb == "ver":
        for case in ver_preterito:
            print(case)

    # verbs with -uv preterite stems (estar, tener, andar)
    if verb in ("estar", "tener", "andar"):
        if verb == "estar":
            for suffix in estar_tener_andar_verb:
                print(verb[:-2] + suffix)
                time.sleep(1)
        elif verb == "tener":
            for suffix in estar_tener_andar_verb:
                print(verb[0] + suffix)
                time.sleep(1)
        elif verb == "andar":
            for suffix in estar_tener_andar_verb:
                print(verb[:-2] + suffix)
                time.sleep(1)

    # REGULAR VERBS
    # verbs that end with -ar
    if verb.endswith("ar"):
        if verb[-3] == "c":
            # replace "c" with "qu" in the first person singular
            conjugation = verb[:-3] + "qu" + ar_suffixes[0]
            print(conjugation)
            time.sleep(1)
            # continue with the regular suffixes for other persons
            for suffix in ar_suffixes[1:]:
                print(verb[:-2] + suffix)
                time.sleep(1)
        elif verb[-3] == "u":
            print(verb[:-3] + "ü" + ar_suffixes[0])
            time.sleep(1)
            for suffix in ar_suffixes[1:]:
                print(verb[:-2] + suffix)
                time.sleep(1)
        elif verb[-3] == "g":
            conjugation = verb[:-3] + "gu" + ar_suffixes[0]
            print(conjugation)
            time.sleep(1)
            for suffix in ar_suffixes[1:]:
                print(verb[:-2] + suffix)
                time.sleep(1)
        elif verb[-3] == "z":
            conjugation = verb[:-3] + "c" + ar_suffixes[0]
            print(conjugation)
            time.sleep(1)
            for suffix in ar_suffixes[1:]:
                print(verb[:-2] + suffix)
                time.sleep(1)

        else:
            for suffix in ar_suffixes:
                print(verb[:-2] + suffix)
                time.sleep(1)

    # er/ ir verbs
    # verbs with u preterite stems (e.g., caber, poner, saber)
    if verb in ("caber", "poner", "saber"):
        if verb == "caber":
            for suffix in caber_saber_poner_hacer_querer_decir_suf:
                print("cup" + suffix)
                time.sleep(1)
        elif verb == "saber":
            for suffix in caber_saber_poner_hacer_querer_decir_suf:
                print("sup" + suffix)
                time.sleep(1)
        elif verb == "poner":
            for suffix in caber_saber_poner_hacer_querer_decir_suf:
                print("pus" + suffix)
                time.sleep(1)
    elif verb in ("hacer", "querer"):
        if verb == "hacer":
            for suffix in caber_saber_poner_hacer_querer_decir_suf[:2]:
                print("hic" + suffix)
                time.sleep(1)
            print("hiz" + caber_saber_poner_hacer_querer_decir_suf[2])
            time.sleep(1)
            for suffix in caber_saber_poner_hacer_querer_decir_suf[3:]:
                print("hic" + suffix)
                time.sleep(1)
        elif verb == "querer":
            for suffix in caber_saber_poner_hacer_querer_decir_suf:
                print("quis" + suffix)
                time.sleep(1)
    elif verb.endswith(("caer", "eer", "oer", "oír", "oir", "uir", "uír")):
        for suffix in er_ir_suffixes[:2]:
            print(verb[:-2] + suffix)
            time.sleep(1)
        print(verb[:-2] + "y" + er_ir_suffixes[2][1])
        time.sleep(1)
        for suffix in er_ir_suffixes[3:-1]:
            print(verb[:-2] + suffix)
            time.sleep(1)
        print(verb[:-2] + "y" + er_ir_suffixes[-1][1:])
        time.sleep(1)
    elif verb.endswith(("er", "ir")):
        # decir
        if verb == "decir":
            for suffix in caber_saber_poner_hacer_querer_decir_suf:
                conjugation = str(("dic" + suffix)).replace("c", "j")
                print(conjugation)
                time.sleep(1)
        elif verb == "venir":
            for suffix in er_ir_suffixes:
                print("vin" + suffix)
                time.sleep(1)
        elif verb[1] == "e":  # verbs like sentir
            for suffix in er_ir_suffixes[:2]:
                print(verb[:-2] + suffix)
                time.sleep(1)
            print(verb[0] + "i" + verb[2:-2] + er_ir_suffixes[2])
            time.sleep(1)
            for suffix in er_ir_suffixes[3:-1]:
                print(verb[:-2] + suffix)
                time.sleep(1)
            print(verb[0] + "i" + verb[2:-2] + er_ir_suffixes[-1])
            time.sleep(1)
        elif verb[1] == "o":  # verbs like dormir, morir
            for suffix in er_ir_suffixes[:2]:
                print(verb[:-2] + suffix)
                time.sleep(1)
            print(verb[0] + "u" + verb[2:-2] + er_ir_suffixes[2])
            time.sleep(1)
            for suffix in er_ir_suffixes[3:-1]:
                print(verb[:-2] + suffix)
                time.sleep(1)
            print(verb[0] + "u" + verb[2:-2] + er_ir_suffixes[-1])
            time.sleep(1)
        elif verb.endswith("ducir"):  # verbs like traducir
            for suffix in er_ir_suffixes[:2]:
                print(str(verb[:-2] + suffix).replace("c", "j"))
                time.sleep(1)
            print(str(verb[:-2] + er_ir_suffixes[2]).replace("ci", "j"))
            time.sleep(1)
            for suffix in er_ir_suffixes[3:-1]:
                print(str(verb[:-2] + suffix).replace("c", "j"))
                time.sleep(1)
            print(str(verb[:-2] + er_ir_suffixes[-1]).replace("ci", "j"))
            time.sleep(1)
        else:
            for suffix in er_ir_suffixes:
                print(verb[:-2] + suffix)
                time.sleep(1)
```

```python
preterito()
```
