## Use

The ***condicional compuesto*** in Spanish is used to express the following:

- the demand or desire of an action that has not taken place.
    
        e.g., Habrías podido avisarnos.

- express an assumption of something that happened in the past.

        e.g., Crees que habríamos dicho que no.
        
- describing the hypothetical scenarion of a past action in conditional sentences.

        e.g., Si ayer no hubiera llovioo, yo y Juan habríamos ido a la playa.
        
        
## Conjugation

To conjugate a Spanish verb in *futuro compuesto* the auxiliary verb ***haber*** in *condicional simple* and the ***participio*** of the main verb are being used.

Verb *haber* in *futuro simple*:

| **Persona** | **Verbo auxiliar** | **Patricipio** |
| --- | --- | --- |
| yo | habría | participio|
| tú | habríás | participio |
| él, ella, usted | habría | participio |
| nosotros/-as | habríamos | participio |
| vosotros/-as | habríais | participio |
| ellos/-as, ustedes| habrían | participio |


### Form of the *participio*
The *participio* (i.e., the past participle) in Spanish is formed by removing the suffix from the verbs (i.e., -ar, -er, -ir) and by adding ***-ado*** (for -ar verbs) or ***-ido*** (for -er or -ir verbs). In other words, by adding the aforementioned suffixed in the root of the infinitive.

    e.g., jugar --> jug-ado
    
    leer --> le-ído
    
    oír --> oído
    
#### Irregular verbs

```python
def futuro_compuesto():
    infinitivo = str(input("Darme un infinitivo en Castellano: ")).lower().strip()

    # auxiliary verb "haber"
    haber_imperf = ["habría", "habrías", "habría", "habríamos", "habríais", "habrían"]
    irreg_verbs = [
        "abrir",
        "decir",
        "escribir",
        "freír",
        "proveer" "hacer",
        "ver",
        "morir",
        "poner",
        "romper",
        "imprimir",
        "satisfacer",
        "suscribir",
        "volver",
    ]

    if infinitivo not in irreg_verbs:
        if infinitivo.endswith("ar"):
            participio = infinitivo[:-2] + "ado"
        elif infinitivo.endswith(("er", "ir")):
            participio = infinitivo[:-2] + "ido"
        elif infinitivo == "ir":
            participio = infinitivo[:-1] + "do"
    if infinitivo == "abrir":
        for aux in haber_imperf:
            print(aux + " " + "abierto")
    elif infinitivo == "decir":
        for aux in haber_imperf:
            print(aux + " " + "dicho")
    elif infinitivo == "escribir":
        for aux in haber_imperf:
            print(aux + " " + "escrito")
    elif infinitivo == "hacer":
        for aux in haber_imperf:
            print(aux + " " + "hecho")
    elif infinitivo == "ver":
        for aux in haber_imperf:
            print(aux + " " + "visto")
    elif infinitivo == "morir":
        for aux in haber_imperf:
            print(aux + " " + "muerto")
    elif infinitivo == "poner":
        for aux in haber_imperf:
            print(aux + " " + "puesto")
    elif infinitivo == "romper":
        for aux in haber_imperf:
            print(aux + " " + "roto")
    elif infinitivo == "freír":
        for aux in haber_imperf:
            print(aux + " " + "frito/freido")
    elif infinitivo == "proveer":
        for aux in haber_imperf:
            print(aux + " " + "provisto/proveido")
    elif infinitivo == "imprimir":
        for aux in haber_imperf:
            print(aux + " " + "impreso/imprimido")
    elif infinitivo == "satisfacer":
        for aux in haber_imperf:
            print(aux + " " + "satisfecho")
    elif infinitivo == "suscribir":
        for aux in haber_imperf:
            print(aux + " " + "suscrito/suscripto")
    elif infinitivo == "volver":
        for aux in haber_imperf:
            print(aux + " " + "vuelto")
    else:
        for aux in haber_imperf:
            print(aux + " " + participio)
```

```python
futuro_compuesto()
```
